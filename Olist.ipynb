```python
db = pd.merge(orders_dataset[['order_id','order_status','order_approved_at']],order_items[['order_id','seller_id','product_id','price']],on= 'order_id',how= 'left')
db = db.loc[db['order_status'] == 'delivered']
db['order_approved_at'] = pd.to_datetime(db['order_approved_at'])
```

```python
safra1 = db[(db['order_approved_at'] >='2016-10-01') & (db['order_approved_at'] <='2017-04-01')]
safra1['mes_por_seller'] = safra1['order_approved_at'].dt.month
safra1['ano_por_seller'] = safra1['order_approved_at'].dt.year
receita_vendedor = safra1.groupby('seller_id').sum('price')
valores_unicos = safra1.groupby('seller_id').nunique()
contagem_valores = safra1.groupby('seller_id').count()
dados_vendedor = pd.merge(receita_vendedor,valores_unicos['order_id'], on = 'seller_id', how= 'right')
dados_vendedor['avg_vl_venda'] = dados_vendedor['price'] / dados_vendedor['order_id']
dados_vendedor = pd.merge(dados_vendedor,contagem_valores['product_id'], on =  'seller_id', how = 'right')
dados_vendedor = pd.merge(dados_vendedor,valores_unicos['product_id'], on =  'seller_id', how = 'right')
dados_vendedor['avg_vl_produto'] = dados_vendedor['price'] / dados_vendedor['product_id_x']
dados_vendedor.rename(columns= {'price':'receita_total','product_id_x':'qte_produtos','product_id_y':'qte_produtos_dst'}, inplace=True)
dados_vendedor['avg_qte_produtos_venda'] = dados_vendedor['qte_produtos'] / dados_vendedor['order_id']
```

```python
primeira_venda = safra1[safra1['order_approved_at'] < '2017-04-01']
primeira_venda = primeira_venda.groupby('seller_id').min() 
ultima_venda = primeira_venda.groupby('seller_id').max() 
d0 = pd.to_datetime('20170401', format='%Y%m%d', errors='ignore')
pd.Timestamp.to_datetime64(d0)
primeira_venda['order_approved_at'] = pd.to_datetime(primeira_venda['order_approved_at'])
primeira_venda['idade_base']= (d0) - (primeira_venda['order_approved_at'])
```

```python
dados_vendedor = pd.merge(dados_vendedor,primeira_venda['idade_base'], on= 'seller_id', how= 'left')
dados_vendedor['idade_base'] = dados_vendedor['idade_base'].dt.days
```

```python
dados_vendedor['idade_base'] = dados_vendedor['idade_base'].astype('int')
dados_vendedor['idade_base_mes'] = (dados_vendedor['idade_base'] / (30))
dados_vendedor.idade_base_mes = dados_vendedor.idade_base_mes.round().astype('int')
```

```python
dados_vendedor['avg_vl_venda_base'] = dados_vendedor['receita_total'] / dados_vendedor['idade_base_mes']
dados_vendedor = dados_vendedor.drop(['mes_por_seller'], axis=1)
dados_vendedor = dados_vendedor.drop(['ano_por_seller'], axis=1)
```

```python
dados_vendedor = pd.merge(dados_vendedor,valores_unicos['mes_por_seller'], on='seller_id', how='left')
dados_vendedor = dados_vendedor[dados_vendedor.replace([np.inf, -np.inf], np.nan).notnull().all(axis=1)]
```

```python
dados_vendedor['prop_ativacao'] = dados_vendedor['idade_base_mes'] / dados_vendedor['mes_por_seller']
dados_vendedor.head(1)
```
